```
# install nnUNet
!git clone https://github.com/MIC-DKFZ/nnUNet.git
!pip install -r nnUNet/requirements.txt
!pip install -e ./nnUNet

# download dataset from https://drive.google.com/open?id=1A2IU8Sgea1h3fYLpYtFb2v7NYdMjvEhU
file_id = '1A2IU8Sgea1h3fYLpYtFb2v7NYdMjvEhU'

from google.colab import auth
auth.authenticate_user()
from googleapiclient.discovery import build
drive_service = build('drive', 'v3')

import io
from googleapiclient.http import MediaIoBaseDownload

request = drive_service.files().get_media(fileId=file_id)
downloaded = io.BytesIO()
downloader = MediaIoBaseDownload(downloaded, request)

downloadCounter = 0
done = False
while done is False:
  status, done = downloader.next_chunk()
  downloadCounter += 1
  if downloadCounter % 100 == 0:
    print('download iteration {}'.format(downloadCounter))
  if status:
    print("Downloading  %d%%." %(int(status.progress() * 100)))
print('Download Complete')


```
